```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import fft
```

## Fourier Analysis

# construct a complex sine wave & plot

```python
n = 1000;
x = np.linspace(0,2*np.pi,n);

#y1 = 2*np.cos(1*x) + 3*np.sin(6*x) - 8*np.sin(11*x) + 2.5*np.cos(23*x)
y1 = 7*np.sin(3*x) + 3*np.cos(7*x) - 2*np.cos(13*x) + 0.25*np.sin(17*x) + 0.1*np.cos(24*x)


a = np.zeros(25); b= np.zeros(25);
for j in range(0,25):

  a[j] = 1/np.pi*np.trapz(y1*np.cos(j*x),x);
  b[j] = 1/np.pi*np.trapz(y1*np.sin(j*x),x);

#print(a)
#print(b)

k = range(0,25);
plt.figure(1);plt.plot(x,y1);
plt.figure(2)
plt.plot(k,a,'o-',k,b,'o-'); plt.legend(['cos coeff','sin coeff']);plt.grid()
```
